```python
#Import packages
import pandas as pd
import numpy as np
import re
```

```python
#Membaca CSV hasil crawling menjadi DataFrame
df = pd.read_csv("https://raw.githubusercontent.com/RandyTunru/PersonalDatasets/master/Tugas%20Akhir/Jakarta%20(6).csv")
```

```python
df.head()
```

```python
#Membuat DataFrame baru dengan kolom "full_text", yang diisi dengan kolom "full_text" dari DataFrame original
new_df = {
    "full_text" : df.full_text,
}

new_df = pd.DataFrame(new_df)

new_df.head()
```

```python
def remove_urls(text):
    # Define the regex pattern to match URLs
    url_pattern = r'https?://\S+|www\.\S+'

    # Use the sub() function to replace URLs with an empty string
    return re.sub(url_pattern, '', text)
```

```python
def remove_tags(text):
    # Define the regex pattern to match tags
    tags_pattern = r'#\S+'

    # Use the sub() function to replace tags with an empty string
    return re.sub(tags_pattern, '', text)
```

```python
def remove_mentions(text):
    # Define the regex pattern to match mentions
    mention_pattern = r'@\S+'

    # Use the sub() function to replace mentions with an empty string
    return re.sub(mention_pattern, '', text)
```

```python
new_df.full_text
```

```python
# Mengeksekusi fungsi ke setiap baris dataframe yang ada
new_df = new_df.full_text.apply(remove_urls).apply(remove_tags).apply(remove_mentions)
```

```python
# Menyimpan DataFrame baru menjadi CSV, dengan nama yang diberikan di parameter
new_df.to_csv("cleaned_tweets.csv")
```
